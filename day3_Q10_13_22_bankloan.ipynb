##### Q10: Using the bankloan.csv data frame, build a column chart that represents the average balance by housing loan category (housing) for the married customers.

```python
#import pandas library for data analysis
#import seaborn and matplotlib library for data visualization

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading the csv files
df = pd.read_csv('bankloan.csv')
df.head()
```

```python
# filtering the married customers
df_married = df[df['marital']== 'married']
df_married
```

```python
average_balance_by_housing = df_married.groupby(by= 'housing')['balance'].mean()
average_balance_by_housing
```

```python
plt.bar(x= average_balance_by_housing.index, height= average_balance_by_housing.values)
plt.title('Average Balance by Housing Loan category for Married Customers')
plt.xlabel('Housing loan category')
plt.ylabel('Average Balance')
plt.show()
```

##### Q13: Using the bankloan.csv data frame, build a mean plot chart that represents the average duration by education level, for the customers who have a personal loan (loan).

```python
#filtering the customers who have a personal loan
df_personal_loan = df[df['loan']== 'yes']
df_personal_loan
```

```python
#groupby education levels
average_duration_by_education = df_personal_loan.groupby('education')['duration'].mean()
average_duration_by_education
```

```python
#plotting a mean plot chart
plt.bar(x= average_duration_by_education.index, height= average_duration_by_education.values)
plt.title('Average Duration by Education level for customers with personal loan')
plt.xlabel('Education level of customers')
plt.ylabel('Average duration')
plt.show()
```
